```
import os
import shutil

dataset_path = '/content/drive/MyDrive/vin/DeepLearning/final_project/my_dataset/test_processed'

labels = ['idle', 'walk', 'wave']
for label in labels:
    path = dataset_path + '/' + label
    for filename in os.listdir(path):
        file_path = path + "/" + filename
        shutil.copyfile(file_path, dataset_path+'/'+filename)
```
