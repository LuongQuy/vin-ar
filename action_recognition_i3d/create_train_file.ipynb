```
import os
import cv2

dataset_path = '/content/drive/MyDrive/vin/DeepLearning/final_project/my_dataset/processed/'
k = 0
train_file = open('/content/drive/MyDrive/vin/DeepLearning/final_project/my_dataset/train_file.txt', 'w')
labels = ['idle', 'walk', 'wave']
cnt = 0
for label in labels:
    path = dataset_path + '/' + label
    for filename in os.listdir(path):
        new_path = path + "/" + filename
        video = cv2.VideoCapture(new_path)
        frame_total = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        if frame_total > 65:
          train_file.write(filename + " " + str(frame_total) + " " + str(k) + "\n")
        cnt += 1
        print(str(cnt) + ':' + new_path, '-> Done')
    k += 1     

train_file.close()
print('Done')
```
