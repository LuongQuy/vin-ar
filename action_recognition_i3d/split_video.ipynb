```
import os
from moviepy.editor import VideoFileClip
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip

dataset_path = '/content/drive/MyDrive/vin/DeepLearning/final_project/my_dataset/test/'
output = '/content/drive/MyDrive/vin/DeepLearning/final_project/my_dataset/test_processed/'
labels = ['idle', 'walk', 'wave']
clip_duration = 3

for label in labels:
    path = dataset_path + label
    k = 0
    for filename in os.listdir(path):
        file_path = path + "/" + filename
        
        clip = VideoFileClip(file_path)
        duration = clip.duration
        i = 0
        while True:
            if i > duration-1:
                break
            new_filename = output+label+'/'+filename+'_'+str(i)+".mp4"
            ffmpeg_extract_subclip(file_path, i, i+clip_duration, targetname=new_filename)
            i += clip_duration
            print(new_filename,'-> Done')
            
```
