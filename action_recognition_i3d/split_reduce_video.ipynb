```
import os
import cv2

dataset_path = '/content/drive/MyDrive/vin/DeepLearning/final_project/dataset/train/'
output = '/content/drive/MyDrive/vin/DeepLearning/final_project/dataset/processed/low_quality/'
labels = ['idle', 'walk', 'wave']
fourcc = cv2.VideoWriter_fourcc(*'MJPG')  # video format
max_frame = 70

reduce_width = 360
reduce_height = 640
for label in labels:
    path = dataset_path + label
    k = 0
    for filename in os.listdir(path):
        new_path = path + "/" + filename
        opencv_video = cv2.VideoCapture(new_path)
        video_writer = None
        cnt = 0
        while True:
            ret, frame = opencv_video.read()
            if not ret:
                break
            fps = opencv_video.get(cv2.CAP_PROP_FPS)
            if video_writer is None:
                new_filename = output + label + '/' + filename + '_' + str(k) + '_output.avi'
                video_writer = cv2.VideoWriter(new_filename, fourcc, fps, (frame.shape[1], frame.shape[0]))
            frame = cv2.resize(frame, (reduce_height, reduce_width)) 
            video_writer.write(frame)
            cnt += 1
            if cnt == max_frame:
                video_writer.release()
                k += 1
                cnt = 0
                video_writer = None
                print(new_filename,'-> Done')
```
