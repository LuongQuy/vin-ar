```python
# Setting up the environment
import numpy as np
import tensorflow as tf
import glob
from os.path import basename
import pandas as pd
import random

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation,\
Flatten, Dense, BatchNormalization, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import utils
```

```python
# Hyper parameters
STEP = 1
N_EPOCH = 2
BATCH_SIZE = 128
VERBOSE = 1
OPTIMIZER = Adam()
NB_CLASSES = 3
input_shape = (32, 12, 2)
```

```python
# load train data
X_train = np.load(f'data_processing/data/train_data/X_train_{STEP}.npy')
y_train = np.load(f'data_processing/data/train_data/y_train_{STEP}.npy')
y_train = utils.to_categorical(y_train, NB_CLASSES)
```

```python
# load val data
X_val = np.load(f'data_processing/data/train_data/X_val_{STEP}.npy')
y_val = np.load(f'data_processing/data/train_data/y_val_{STEP}.npy')
y_val = utils.to_categorical(y_val, NB_CLASSES)
```

```python
def get_ehpi_model():
    model = Sequential()
    model.add(Conv2D(64, kernel_size=3, padding="same", input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(Conv2D(64, kernel_size=3, padding="same"))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(MaxPooling2D())

    model.add(Conv2D(128, kernel_size=3, padding="same"))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(Conv2D(128, kernel_size=3, padding="same"))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(MaxPooling2D())

    model.add(Conv2D(256, kernel_size=3, padding="same"))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(Conv2D(256, kernel_size=3, padding="same"))
    model.add(BatchNormalization())
    model.add(Activation("relu"))

    model.add(GlobalAveragePooling2D())
    model.add(Dense(3))
    model.add(Activation("softmax"))
    return model
```

```python
model = get_ehpi_model()
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer=OPTIMIZER, metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=BATCH_SIZE, epochs=N_EPOCH, verbose=VERBOSE, validation_data=(X_val, y_val))
```

```python
model.save('demo/test.h5')
```
