```
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix
```

```
X = np.load('data_processing/data/test_data/test_poses_1.npy')
```

```
y = np.load('data_processing/data/test_data/test_labels_1.npy')
```

```
model = tf.keras.models.load_model('demo/test.h5')
```

```
y_pred = model.predict(X)
```

```
y_pred = np.argmax(y_pred, axis=1)
```

```
print(np.count_nonzero(y_pred == y) / 6000)
```

```
confusion_matrix(y, y_pred)
```
